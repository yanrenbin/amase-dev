```python
import numpy
import math
import matplotlib.pyplot as plt
```

```python
angle=np.arange(37)*10
dtor = math.pi/180.
rr=50./2.
spacing=100.
xx=rr*np.cos(angle*dtor)
yy=rr*np.sin(angle*dtor)
n0 = 37
for i in np.arange(37)-18:
    x0 = np.arange(n0-i)*spacing+spacing/2*i
    y0 = spacing*i*np.sqrt(3.)/2.+np.zeros(n0-i)
    for j in range(n0-i):
        plt.plot(x0[j]+xx,y0[j]+yy,color='blue')
plt.show()
```
