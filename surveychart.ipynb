```python
import numpy as np
import matplotlib.pyplot as plt
import math

arcinrad = 206265.
amasefocal = 400 #mm
fiberdia = 0.05 #mm
amase_angres = 206265/amasefocal*fiberdia
nfibers = 1027.
amase_angcov = np.sqrt(nfibers)*amase_angres#in arcsec. This is the effective diameter of the actual coverage of all the fibers. It is the same as firt calculating the total covered area of all the fibers, then converting that to the diameter of the circle with the same area. 
amase_angcov_arr = amase_angcov*10.

manga_angres = 2.5
manga_angcov = np.sqrt(127.)*2.0
lvm_angres = arcinrad/(161*11.42)*0.330
lvm_angcov = np.sqrt(1801)*lvm_angres

sami_angres = np.sqrt(1.6**2+1.0**2)
sami_angcov = np.sqrt(61.)*1.6
sami_res = [1808,4304]

kcwi_angres_s = np.sqrt(0.35**2+0.7**2)
kcwi_angcov_s = 2*np.sqrt(8*20./math.pi)
kcwi_angres_m = np.sqrt(0.7**2+0.7**2)
kcwi_angcov_m = 2*np.sqrt(16*20/math.pi)
kcwi_angres_l = 1.35
kcwi_angcov_l = 2*np.sqrt(33*20/math.pi)


muse_angres = np.sqrt(0.6**2+0.35**2)
muse_angcov = np.sqrt(60.*60./math.pi)*2.

ppak_angres = 2.7
ppak_angcov = np.sqrt(331)*2.7
ppak_res = [870, 1400, 1778,2566, 3780, 4918, 16111]

sparsepak_angres=4.687
sparsepak_angcov=np.sqrt(82.)*4.687
sparsepak_res = [4818, 10049, 20318]

weave_lifu_angres=2.6
weave_angres = 1.3
weave_lifu_angcov = np.sqrt(547.)*2.6
weave_res = [5750,21000]

wham_angres = 3600.
wham_angcov = 2*3600.

hetdex_angres = 1.5
hetdex_angcov = np.sqrt(35000)*1.5

mw_low = 0.4  #kpc
mw_hi = 3.0  #kpc

lmc = 50. #kpc
m31 = 780.

tenmpc = 10000

virgo = 16000

hundredmpc = 100000

amase_res = 15000
kcwi_res_s = [3600, 8000, 18000]
kcwi_res_m = [1800, 4000, 9000]
kcwi_res_l = [900,2000,4500]
muse_res = 3000
lvm_res = 3500
manga_res = 2000
wham_res = 30000
hetdex_res = 750

fig2 = plt.figure(figsize=(6,4),dpi=100)
ax2 = fig2.add_subplot(111)
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_ylim([500,35000])
ax2.set_xlabel('Spatial Resolution and Spatial Coverage (arcsec)', fontsize=14)
ax2.set_ylabel('Spectral Resolution (R)', fontsize=14)
ax2.plot([amase_angres,amase_angcov],[amase_res,amase_res],linewidth=6,color='C0')
ax2.text(amase_angres*4,amase_res*1.15,'LAHRSS',color='C0', fontweight='bold')
#ax2.plot([amase_angres,amase_angcov_arr],[amase_res,amase_res],linewidth=2,color='C0')
#ax2.text(amase_angcov*2,amase_res*1.15,'AMASE',color='C0', fontweight='bold')
ax2.plot([lvm_angres,lvm_angcov],[lvm_res,lvm_res],linewidth=6,color='C1')
ax2.text(lvm_angres*2,lvm_res*1.2,'SDSS-V/LVM',color='C1',fontweight='bold')
ax2.plot([sami_angres,sami_angcov],[sami_res,sami_res],linewidth=6,color='C9')
ax2.text(sami_angres,sami_res[-1]*1.1,'SAMI',color='C9',fontweight='bold')
ax2.plot([manga_angres,manga_angcov],[manga_res,manga_res],linewidth=6,color='C2')
ax2.text(manga_angres*2,manga_res*1.1,'MaNGA',color='C2', fontweight='bold')
ax2.plot([muse_angres,muse_angcov],[muse_res,muse_res],linewidth=6,color='C3')
ax2.text(muse_angres*1.3,muse_res*0.8,'MUSE',color='C3', fontweight='bold')
ax2.plot([ppak_angres,ppak_angcov],[ppak_res,ppak_res],linewidth=2,color='C8')
ax2.text(ppak_angres*1.5,ppak_res[-1]*0.8,'PPAK',color='C8',fontweight='bold')
ax2.plot([weave_angres,weave_lifu_angcov],[weave_res,weave_res],linewidth=2,color='C4')
ax2.text(weave_angres*2,weave_res[-1]*1.1,'WEAVE IFU',color='C4',fontweight='bold')
ax2.plot([sparsepak_angres,sparsepak_angcov],[sparsepak_res,sparsepak_res],linewidth=2,color='C5')
sparsepak_angmed = np.sqrt(sparsepak_angres*sparsepak_angcov)
ax2.text(sparsepak_angres*1.5,sparsepak_res[-2]*1.1,'SparsePak',color='C5', fontweight='bold')
for res in kcwi_res_s:
    ax2.plot([kcwi_angres_s,kcwi_angcov_s],[res,res],linewidth=2,color='C6')
for res in kcwi_res_m:
    ax2.plot([kcwi_angres_m,kcwi_angcov_m],[res,res],linewidth=2,color='C6')
for res in kcwi_res_l:
    ax2.plot([kcwi_angres_l,kcwi_angcov_l],[res,res],linewidth=2,color='C6')
ax2.text(kcwi_angres_s*1.3,kcwi_res_s[2]*0.8,'KCWI',color='C6', fontweight='bold')
ax2.plot([hetdex_angres,hetdex_angcov],[hetdex_res,hetdex_res],linewidth=6,color='C7')
ax2.text(hetdex_angres*48,hetdex_res*1.2,'HETDEX',color='C7', fontweight='bold')

def arcsec2pc(x):
    return x*2000/206265.

def pc2arcsec(x):
    return x*206265/2000.

secax = ax2.secondary_xaxis('top', functions=(arcsec2pc, pc2arcsec))
secax.set_xlabel('Physical Scale at a distance of 2kpc (pc)',fontsize=14)
```
