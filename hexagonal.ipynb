```python
import numpy
import math
import matplotlib.pyplot as plt
```

```python
angle=np.arange(37)*10
dtor = math.pi/180.
rr=50./2.
r2=85./2.
spacing=85.
xx=rr*np.cos(angle*dtor)
yy=rr*np.sin(angle*dtor)
xx2 = r2*np.cos(angle*dtor)
yy2 = r2*np.sin(angle*dtor)
n0 = 37
fig,axs=plt.subplots(1,1,figsize=(20,20))
for i in np.arange(37)-18:
    x0 = np.arange(n0-np.abs(i))*spacing+spacing/2*np.abs(i)
    y0 = spacing*i*np.sqrt(3.)/2.+np.zeros(n0-np.abs(i))
    for j in range(n0-np.abs(i)):
        axs.fill(x0[j]+xx,y0[j]+yy,color=u'#ff7f0e')
        axs.plot(x0[j]+xx2,y0[j]+yy2,color=u'#ff7f0e')
axs.axis('equal')
print(plt.rcParams['axes.prop_cycle'].by_key()['color'])
plt.show()
```
