```python
from astropy.io import ascii
import numpy as np
import matplotlib.pyplot as plt


dir = '/Users/renbin/astro/amase/optical_model/'
defaultslit = ascii.read(dir+'vignettingmodel_flatslit_default.txt')
#curvslit = ascii.read(dir+'vignettingmodel_curvedslit.txt')
rectslit = ascii.read(dir+'vignettingmodel_flatslit_6040.txt')
defaultthrupt=np.vstack([defaultslit['col2'],defaultslit['col3'],defaultslit['col4'],defaultslit['col5']])#,defaultslit['col6'],defaultslit['col7'],defaultslit['col8']])
rectthrupt=np.vstack([rectslit['col2'],rectslit['col3'],rectslit['col4'],rectslit['col5']])
fig,axes = plt.subplots(1,2,dpi=150)
axes[0].imshow(np.transpose(defaultthrupt/np.max(defaultthrupt)),vmin=0,vmax=1,cmap='jet',extent=[-5,35,695,615])
axes[0].invert_yaxis()
axes[0].set_xlabel('Slit Position(mm)')
axes[0].set_ylabel('Wavelength (nm)')
axes[0].set_title('50micron round fiber')
ax1=axes[1].imshow(np.transpose(rectthrupt/np.max(defaultthrupt)),vmin=0,vmax=1,cmap='jet',extent=[-5,35,695,615])
axes[1].invert_yaxis()
axes[1].set_xlabel('Slit Position(mm)')
axes[1].set_title('60*40micron aperture')
cbar = fig.colorbar(ax1, ax=axes.ravel().tolist(), shrink=0.95)
```
